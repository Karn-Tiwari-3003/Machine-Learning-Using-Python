```python
from pydataset import data
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
pima = data('Pima.tr')
```

```python
pima
```

```python
pima.plot(kind='scatter', x = 'skin', y = 'bmi')
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(pima.skin, pima.bmi)
```

```python
plt.scatter(X_train, Y_train, label='Training Data', color='r', alpha = 0.7)
plt.scatter(X_test, Y_test, label='Training Data', color='b', alpha = 0.7)
plt.legend()
plt.title('Train Test Data')
plt.show()
```

```python
LR = LinearRegression()
LR.fit(X_train.values.reshape(-1,1), Y_train)
```

```python
#test sets
prediction = LR.predict(X_test.values.reshape(-1,1))

plt.plot(X_test, prediction, label='Linear Regression', color='r')
plt.scatter(X_test, Y_test, label='Training Data', color='b', alpha = 0.7)
plt.legend()
plt.title('Linear Regression')
plt.show()
```

```python
#traing sets
prediction = LR.predict(X_train.values.reshape(-1,1))

plt.plot(X_train, prediction, label='Linear Regression', color='r')
plt.scatter(X_train, Y_train, label='Training Data', color='b', alpha = 0.7)
plt.legend()
plt.title('Linear Regression')
plt.show()
```

```python
LR.score(X_test.values.reshape(-1,1), Y_test)
```
