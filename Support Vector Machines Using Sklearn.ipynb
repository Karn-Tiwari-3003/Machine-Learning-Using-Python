```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df
```

```python
df['Target'] = iris.target
```

```python
df.head()
```

```python
iris.target_names
```

```python
df[df.Target == 1].head()
```

```python
df[df.Target == 2].shape
```

```python
df[df.Target == 0].shape
```

```python
df['flower_name'] = df.Target.apply(lambda x : iris.target_names[x]) 
```

```python
df.head()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df0 = df[df.Target == 0]
df1 = df[df.Target == 1]
df2 = df[df.Target == 2]
```

```python
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='g', marker='+', label='0')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='b', marker='+', label='1')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color='r', marker='+',label='2')
plt.xlabel('sepal length(cm)')
plt.ylabel('sepal width(cm)')
plt.legend()
plt.show()
```

```python
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color='g', marker='+', label='0')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='b', marker='+', label='1')
plt.scatter(df2['sepal length (cm)'], df2['petal width (cm)'], color='r', marker='+',label='2')
plt.xlabel('petal length(cm)')
plt.ylabel('petal width(cm)')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['Target', 'flower_name'], axis = 1)
```

```python
X.head()
```

```python
y = df.Target
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.1)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(C=1, gamma=1, kernel='linear')
```

```python
model.fit(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```
